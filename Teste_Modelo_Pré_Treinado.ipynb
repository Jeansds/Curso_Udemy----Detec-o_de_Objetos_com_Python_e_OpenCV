```
import cv2
from google.colab.patches import cv2_imshow
```

```
imagem = cv2.imread('/content/sample_data/pessoas.jpg')
classificador = cv2.CascadeClassifier('/content/sample_data/haarcascade_frontalface_default.xml')#classificador p´re pronto
imagemcinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
deteccoes = classificador.detectMultiScale(imagemcinza, scaleFactor = 1.1,
                                           minNeighbors= 5, minSize = (30, 30),
                                           maxSize = (120, 120))
```

```
print(deteccoes)
print(len(deteccoes))
```

```
for (x, l, y, a) in deteccoes:#l = largura, a = altura, x = x inicial do eixo, y = y inicial do eixo
  cv2.rectangle(imagem, (x, y), (x + l, y + a), (0, 255, 0), 2)
cv2_imshow(imagem)
```
